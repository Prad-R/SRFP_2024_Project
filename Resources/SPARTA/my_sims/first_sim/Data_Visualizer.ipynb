```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import re
```

```python
part_10ppgc = {}

directory = 'grid_files/10ppgc_results'

part_files = []

for filename in os.listdir(directory):
    if re.match('particles.*.grid', filename):
        part_files.append(os.path.join(directory, filename))

for part_file in part_files:
    part_file_index = int(re.findall('[0-9]{1,9}', part_file)[1])

    with open(part_file) as f:
        j = 1
        temp = {}

        for line in f:
            line = line.replace(' \n', '', 1)
            temp[j] = line
            j+=1

    part_10ppgc[part_file_index] = temp

part_10ppgc = dict(sorted(part_10ppgc.items()))
```

```python
with open('grid_files/10ppgc_results/particles.csv', 'w') as f:
    csvfile = csv.DictWriter(f, fieldnames = )

    for key in part_10ppgc.keys():
        csvfile.writerow(part_10ppgc[key])
```
