```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
columns = ['VL', 'Day', 'T_Mean', 'T_Min', 'T_Range', 'Flag']
df = pd.read_csv("VL_Avg_Temp_Data.csv", usecols = [0, 3, 4, 5, 6, 15], names = columns)

df = df.drop(df[df['Flag'] != "BB"].index)
df = df.drop(df[df['VL'] != "VL1"].index)
```

```python
days = np.arange(min(df['Day']), max(df['Day']), 1)
T_avg = np.average(df['T_Mean']) * np.ones(len(days))

plt.figure(figsize = (10, 7.5))
plt.scatter(df['Day'], df['T_Mean'], s = 5, label = "Average Footpad Temperature")
plt.plot(days, T_avg, color = 'r', label = "Time Average Surface Temperature")
plt.grid()
plt.xlabel("Martian Day $\\longrightarrow$", fontsize = 15)
plt.ylabel("Temperature (in K) $\\longrightarrow$", fontsize = 15)
plt.legend(fontsize = 15)
plt.title("Temperature on Surface", fontsize = 21)
plt.savefig("Surface_Temp.png")
```
